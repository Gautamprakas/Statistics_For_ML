```python
'''Q2: Differentiate between one-tailed and two-tailed tests.
Answer-A one-tailed test and a two-tailed test are two types of statistical hypothesis tests used to determine the statistical 
significance of a sample mean.

In a one-tailed test, the alternative hypothesis specifies the direction of the difference between the sample mean and the
population mean. The researcher is interested in whether the sample mean is either significantly greater than or significantly
less than the population mean. Therefore, the rejection region is only on one side of the distribution. A one-tailed test is
used when there is prior knowledge or a strong theoretical reason to expect the direction of the difference.

In contrast, a two-tailed test does not specify the direction of the difference. The researcher is interested in whether 
the sample mean is significantly different from the population mean, regardless of whether it is greater or less than the
population mean. Therefore, the rejection region is divided into two equal parts, one in each tail of the distribution.
A two-tailed test is used when there is no prior knowledge or theoretical reason to expect the direction of the difference.

For example, suppose a researcher wants to test whether a new teaching method improves test scores. A one-tailed test could 
be used if the researcher hypothesizes that the new method will increase test scores, and they want to see if the mean test
score is significantly greater than the population mean. A two-tailed test could be used if the researcher has no prior
knowledge about the direction of the effect and wants to see if the mean test score is significantly different from the 
population mean.'''
```

```python
'''Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
Answer-In hypothesis testing, there are two types of errors: Type 1 error and Type 2 error.

Type 1 error occurs when the null hypothesis is rejected when it is actually true. It is also known as a false positive error. 
The probability of making a Type 1 error is denoted by alpha (α) and is typically set at 0.05 or 0.01. Type 1 errors occur 
when the researcher concludes that there is a significant difference between the sample mean and the population mean, even 
though the difference is due to random chance.

An example of Type 1 error is when a medical researcher concludes that a new drug is effective in treating a disease when, 
in fact, it is not. This could happen if the researcher's hypothesis testing procedure mistakenly concludes that there is a
significant difference between the treatment and placebo groups, even though the difference is simply due to chance.

Type 2 error occurs when the null hypothesis is not rejected when it is actually false. It is also known as a false negative
error. The probability of making a Type 2 error is denoted by beta (β). Type 2 errors occur when the researcher fails to
detect a significant difference between the sample mean and the population mean, even though there is a real difference.

An example of Type 2 error is when a researcher fails to reject the null hypothesis that a new treatment has no effect on a
disease when, in fact, it does. This could happen if the study has a small sample size or a low statistical power, making it
difficult to detect a true difference between the sample mean and the population mean.

In summary, Type 1 error occurs when a researcher rejects a true null hypothesis, while Type 2 error occurs when a researcher
fails to reject a false null hypothesis. Both types of errors can have serious consequences and can lead to incorrect 
conclusions in hypothesis testing.'''
```

```python
'''Q4: Explain Bayes's theorem with an example.
Answer-Bayes's theorem is a statistical rule that describes the probability of an event based on prior knowledge of 
conditions that might be related to the event. It is a way of updating our prior beliefs about the probability of an 
event based on new evidence.

Bayes's theorem states that:

P(A|B) = P(B|A) * P(A) / P(B)

where:

P(A|B) is the conditional probability of event A given that event B has occurred
P(B|A) is the conditional probability of event B given that event A has occurred
P(A) is the prior probability of event A occurring
P(B) is the prior probability of event B occurring
An example of Bayes's theorem is in medical diagnosis. Suppose a test has a 99% accuracy rate for detecting a disease, 
and the prevalence of the disease in the population is 0.1%. A person takes the test and tests positive for the disease.
What is the probability that the person actually has the disease?

Let A be the event that the person has the disease, and B be the event that the person tests positive for the disease.

P(A) = 0.001 (the prior probability of the person having the disease)
P(B|A) = 0.99 (the probability of testing positive given that the person actually has the disease)
P(B|~A) = 0.01 (the probability of testing positive given that the person does not have the disease)
P(~A) = 0.999 (the prior probability of the person not having the disease)

Using Bayes's theorem, we can calculate the probability that the person actually has the disease given that they 
tested positive:

P(A|B) = P(B|A) * P(A) / [P(B|A) * P(A) + P(B|~A) * P(~A)]
= 0.99 * 0.001 / [0.99 * 0.001 + 0.01 * 0.999]
= 0.09

Therefore, even though the person tested positive for the disease, the probability of them actually having the disease is 
only 9%. This shows the importance of taking into account the prevalence of the disease and the accuracy of the test when 
interpreting the results.'''
```

```python
'''Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Answer-A confidence interval is a range of values that is likely to contain the true population parameter with a certain 
level of confidence. It is a statistical measure used to estimate the uncertainty around a sample statistic, such as the 
mean or proportion.

To calculate a confidence interval, we need to know the sample size, sample mean, sample standard deviation 
(or standard error), and the level of confidence. The level of confidence is typically expressed as a percentage
and represents the probability that the true population parameter falls within the calculated interval.

The formula for calculating a confidence interval for the population mean is:

Confidence Interval = sample mean ± (t-value * standard error)

where:

sample mean is the mean of the sample
t-value is the critical value of the t-distribution based on the degrees of freedom and level of confidence
standard error is the standard deviation of the sampling distribution, which is estimated using the sample standard
deviation and sample size
For example, suppose we want to estimate the average height of students in a school with a sample of 50 students.
We measure the height of the sample and find a sample mean of 65 inches and a sample standard deviation of 3 inches.
We want to calculate a 95% confidence interval for the true population mean.

First, we need to calculate the standard error:

standard error = sample standard deviation / sqrt(sample size)
= 3 / sqrt(50)
= 0.424

Next, we need to find the t-value for a 95% confidence level with 49 degrees of freedom (sample size minus 1) 
using a t-distribution table or calculator. Assuming a two-tailed test, the t-value is approximately 2.009.

Finally, we can calculate the confidence interval:

Confidence Interval = sample mean ± (t-value * standard error)
= 65 ± (2.009 * 0.424)
= (64.15, 65.85)

Therefore, we can say with 95% confidence that the true population mean height of students in the school falls between
64.15 and 65.85 inches.'''
```

```python
'''Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
Answer-Suppose there are two machines that produce widgets, Machine A and Machine B. Machine A produces 60% of the widgets, while Machine B produces 40%. However, 5% of the widgets produced by Machine A are defective, while 10% of the widgets produced by Machine B are defective. A widget is selected at random from the production line, and it is found to be defective. What is the probability that it was produced by Machine B?

Let's define the events:

A: the widget was produced by Machine A
B: the widget was produced by Machine B
D: the widget is defective
We want to find P(B|D), the probability that the widget was produced by Machine B given that it is defective. We can use Bayes' Theorem to calculate this probability:

P(B|D) = P(D|B) * P(B) / [P(D|A) * P(A) + P(D|B) * P(B)]

We can plug in the given probabilities:

P(D|A) = 0.05
P(D|B) = 0.1
P(A) = 0.6
P(B) = 0.4

We can now calculate P(B|D):

P(B|D) = 0.1 * 0.4 / [0.05 * 0.6 + 0.1 * 0.4]
= 0.4444

Therefore, the probability that the defective widget was produced by Machine B is approximately 44.44%.'''
```
