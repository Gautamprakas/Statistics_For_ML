```python
'''Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.
Answer-In probability and statistics, a probability mass function is a function that gives the probability that a discrete
random variable is exactly equal to some value. Sometimes it is also known as the discrete density function
## A PMF is a function that describes the probability of each possible outcome of a discrete random variable. 
## It maps each value of the random variable to the probability of that value occurring. 
## The PMF is defined only for discrete random variables, which take on a finite or countably infinite set of possible values.

##   A PDF, on the other hand, is a function that describes the probability distribution of a continuous random variable.
##   It represents the relative likelihood of different values of the random variable occurring within a certain interval. 
##   The PDF is defined for continuous random variables, which can take on any value within a certain range.'''
import numpy as np
import matplotlib.pyplot as plt

# Define a discrete random variable with values 1, 2, 3, 4, and 5
x = np.array([1, 2, 3, 4, 5])

# Define the probabilities of each value occurring
pmf = np.array([0.1, 0.2, 0.3, 0.2, 0.2])

# Plot the PMF
plt.stem(x, pmf)
plt.title('Probability Mass Function (PMF)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()

# Define a continuous random variable with a normal distribution
x = np.linspace(-5, 5, 1000)
pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Plot the PDF
plt.plot(x, pdf)
plt.title('Probability Density Function (PDF)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
'''Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Answer-The CDF is defined for both discrete and continuous random variables. For a discrete random variable, 
 the CDF is defined as the sum of the probabilities of all values less than or equal to x.
For a continuous random variable, the CDF is defined as the integral of the PDF from negative infinity to x.'''
import numpy as np
import matplotlib.pyplot as plt


mu = 0
sigma = 1
x = np.linspace(-5, 5, 1000)
pdf = (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-0.5 * ((x - mu) / sigma)**2)


cdf = np.cumsum(pdf) * (x[1] - x[0])

# Plot the PDF and CDF
fig, ax = plt.subplots()
ax.plot(x, pdf, label='PDF')
ax.plot(x, cdf, label='CDF')
ax.legend()
ax.set_title('Normal Distribution')
ax.set_xlabel('Value')
ax.set_ylabel('Probability')
plt.show()
## The CDF is useful because it allows us to compute the probability of X taking on any value within a certain range.
## We can also use the CDF to compute various statistical measures, such as the median and quartiles of the distribution.
```

```python
'''Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal
Distribution.
Answer-The normal distribution is an important probability distribution in statistics and data analysis.
It has several important properties that make it useful in modeling real-world phenomena, such as:

 Symmetry: The normal distribution is symmetric around its mean, which makes it easy to compute probabilities and statistical measures.

Central limit theorem: The normal distribution is a key component of the central limit theorem,
which states that the distribution of the sum or average of a large number of independent and identically distributed random
variables approaches a normal distribution.

Common occurrence: Many real-world phenomena follow a normal distribution, such as height, weight, and test scores.
example:- height of people in population'''
import numpy as np
import matplotlib.pyplot as plt

# Define a population of heights with mean 170 cm and standard deviation 5 cm
mu = 170
sigma = 5
heights = np.random.normal(mu, sigma, size=10000)

# Plot the distribution of heights
fig, ax = plt.subplots()
ax.hist(heights, bins=50)
ax.set_title('Distribution of Heights')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Frequency')
plt.show()
```

```python
'''Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
Answer-The Bernoulli distribution is a discrete probability distribution that describes the probability of a single binary event with two possible outcomes, typically labeled as success and failure. 
 The distribution is characterized by a single parameter p, which represents the probability of success
 examples:- bernouli distribution'''
import numpy as np
import matplotlib.pyplot as plt

# Define the probability of success
p = 0.5

# Generate a random sample of 1000 trials from a Bernoulli distribution
trials = np.random.binomial(1, p, size=1000)

# Plot the distribution of the sample
fig, ax = plt.subplots()
ax.hist(trials)
ax.set_title('Distribution of Bernoulli Trials')
ax.set_xlabel('Outcome')
ax.set_ylabel('Frequency')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Failure', 'Success'])
plt.show()
## The main difference between the Bernoulli distribution and the binomial distribution is that the Bernoulli distribution models a single trial or event,
## while the binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.
## The binomial distribution is characterized by two parameters: the probability of success p and the number of trials n.
## examples :- Binomial distribution
import numpy as np
import matplotlib.pyplot as plt

# Define the number of trials and the probability of success
n = 10
p = 0.5

# Generate a random sample of 1000 trials from a binomial distribution
trials = np.random.binomial(n, p, size=1000)

# Plot the distribution of the sample
fig, ax = plt.subplots()
ax.hist(trials, bins=range(n+2))
ax.set_title('Distribution of Binomial Trials')
ax.set_xlabel('Number of Successes')
ax.set_ylabel('Frequency')
plt.show()
```

```python
'''Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.'''
from scipy.stats import norm

mu = 50
sigma = 10
x = 60

# Calculate the z-score
z = (x - mu) / sigma

# Calculate the probability using the cumulative distribution function (CDF)
p = 1 - norm.cdf(z)

print(f"The probability of a randomly selected observation being greater than 60 is {p:.4f}")
```

```python
'''Q7: Explain uniform Distribution with an example.'''
## 7 answer
## The uniform distribution is a continuous probability distribution that has a constant probability density function over a specific interval. 
## In other words, all values in the interval have an equal chance of being selected.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the lower and upper boundaries
lower_bound = 0
upper_bound = 10

# Create a uniform distribution object
uniform_dist = uniform(loc=lower_bound, scale=upper_bound-lower_bound)

# Generate random samples from the distribution
samples = uniform_dist.rvs(size=1000)

# Plot the probability density function
x = np.linspace(lower_bound, upper_bound, 100)
pdf = uniform_dist.pdf(x)
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Uniform Distribution')
plt.show()
```

```python
'''Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
Answer-In probability theory, the central limit theorem establishes that, in many situations, for identically distributed 
independent samples, the standardized sample mean tends towards the standard normal distribution even if the original 
variables themselves are not normally distributed.
The central limit theorem is useful when analyzing large data sets because it allows one to assume that the sampling distribution 
of the mean will be normally-distributed in most cases. This allows for easier statistical analysis and inference'''
```
